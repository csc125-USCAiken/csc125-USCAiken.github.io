## Try it ---------------------
Create a program that convert  temperature from Fahrenheit to Celsius. 
Screenshot from 2020-09-10 01-40-31.png


````
# This is a program to convert Fahrenheit to Celsius
# author: Adam Boese
# last updated: 09/21/2020

# Input - Temp in F
tempF = input("What is the temperature in Fahrenheit? \n")

# Process - Deduct 32, Multiply by 5, then divide by 9
tempC = ((float(tempF) - 32) * 5) / 9

# Output 
print(f'{tempF}F converts to {tempC:.2f}C')
````

## Try it -------------------
BMI Calcultor

Create a simple BMI calculator that lets the user enter his or her weight(in pounds)  and height(in inches), then calculates a person’s body mass index.  A person’s BMI is calculated with the following formula: BMI = weightx703/ height^2. 

Screenshot from 2020-09-10 01-41-38.png
 
  


````
# This program is a simple BMI Calculator
# author: Adam Boese
# last updated: 09/21/2020

# input - weight(lb) and height(in)
weight = float(input("Let figure out your BMI!\nWhats your weight in (lbs)? \n"))
height = float(input("Great! Now what's your height in inches? \n"))

# process - bmi=weightX703/height^2
bmi = (weight)*703/(height)**2

# output
print(f'At {weight}lbs and {height}in tall, your BMI is {bmi:.2f}.')
````

````
# This program is a mortage calculator
# author: Adam Boese
# last modified: 09/21/2020

#input - Loan Amount, Interest Rate(annual), Loan Term(years) 
loanAmt = float(input("What is the Loan Amount in $USD? \n"))
apr = float(input("What is the Annual Interest Rate %? \n"))
loanYear = float(input("What is the term of the loan in years? \n"))

#process - Mortgage Calc Formula
#change apr% to decimal and monthly /  loanTerm in years to monthly
mpr = ((apr)/100)/12
loanMonth = (loanYear)*12

monthPay = (loanAmt) * ((mpr)*(1+(mpr))**(loanMonth))/((1+(mpr))**(loanMonth)-1)

#output - Monthly Payment
print(f'Your monthly payment is ${monthPay:.2f}.')
````
