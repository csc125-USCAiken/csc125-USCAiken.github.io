## Universal Sentence Encoder



### Setup and experiment

This section sets up the environment for access to the Universal Sentence Encoder on TF Hub and provides examples of applying the encoder to words, sentences, and paragraphs.

Please run the following code step by step

2. Now run these codes.

```
#@title Load the Universal Sentence Encoder's TF Hub module
from absl import logging

import tensorflow as tf

import tensorflow_hub as hub
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import re
import seaborn as sns

module_url = "https://tfhub.dev/google/universal-sentence-encoder/4" #@param ["https://tfhub.dev/google/universal-sentence-encoder/4", "https://tfhub.dev/google/universal-sentence-encoder-large/5"]
model = hub.load(module_url)
print ("module %s loaded" % module_url)
def embed(input):
  return model(input)
```

3. Now keep runing the following code and perform the experiment. 


*   Enter a sentence that is similar to "My phone is not good".
*   Enter a sentence that is related to "My phone is not good".
*   Enter a sentence that is totally different from "My phone is not good".




```
#Check whether certain messages are similar. The score is from 0 to 1. 

mysentence = "My phone is not good."
yoursentence = input("Please input a similar sentence ")
message_embeddings = embed([mysentence, yoursentence])
score = np.inner(message_embeddings, message_embeddings)
print(f"The similarity between [ {mysentence} ] and [ {yoursentence} ] is {score[0][1]}")
```
